# RAG sur tweets climat avec HuggingFace, LangChain et GPT-2

Ce notebook reprend le pipeline du projet : charger une base de tweets HuggingFace, indexer les tweets dans Chroma, puis questionner GPT-2 avec les tweets retrouves par LangChain.

```python
from climate_tweet_rag import ClimateTweetRAG, load_config, load_tweet_documents, split_documents

config = load_config()
config
```

## 1. Charger quelques tweets

```python
documents = load_tweet_documents(config)
chunks = split_documents(documents, config)

len(documents), len(chunks), documents[0].page_content
```

## 2. Construire l'index Chroma

```python
rag = ClimateTweetRAG(config)
vector_store = rag.build_vector_store(reset=True)
vector_store._collection.count()
```

## 3. Poser une question

```python
question = "What opinions appear in the tweets about climate change?"
answer, sources = rag.answer(question)

print(answer)
print("\nSources:")
for index, doc in enumerate(sources, start=1):
    print(f"\n[{index}] {doc.page_content}")
```
